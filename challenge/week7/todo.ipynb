```python
def add_task(task_list, task): # task_list에 task 추가하기
    task_list.append(task)
    
def display_tasks(task_list): # task_list에 업무들을 반복문을 출력한다.
    if not task_list:
        print('할 일이 없습니다.')
    else:
        for tasks in task_list:
            print(tasks)
    
def save_tasks_to_file(task_list, file_name): # 업무 리스트에 있는 업무들을 반복문을 통해 file_name에 추가한다.
    file_name = open(file_name, 'w', encoding = 'utf8')
    for task in task_list:
        file_name.writelines(task)
    file_name.close()
    
def load_tasks_from_file(file_name): # 업무 리스트를 초기화하고 file_name으로부터 업무 리스트를 불러들여 초기화된 업무 리스트에 추가한다.
    task_list = []
    file_name = open(file_name, 'r', encoding = 'utf8')
    file_task = file_name.readlines()
    task_list.append(file_task)
    file_name.close()
    return task_list

file_name = "task.txt"
task_list = load_tasks_from_file(file_name)

while True:
    print('\n할일 관리 프로그램')
    print('1. 할일 추가')
    print('2. 할일 목록 표시')
    print('3. 종료')
    
    choice = input('선택: ')
    
    if choice == '1':
        task = input('할일을 입력하세요: ')
        add_task(task_list, task)
        print('할일이 추가되었습니다.')
    elif choice == '2':
        display_tasks(task_list)
    elif choice == '3':
        save_tasks_to_file(task_list, 'task.txt')
        print('프로그램을 종료합니다.')
        break
    else:
        print('잘못된 선택입니다. 다시 시도하세요.')
```
